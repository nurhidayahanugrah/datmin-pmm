```python
import numpy as np
import pandas as pd
```

```python
# Nama file Excel
file_names = ['perceraian2018.xlsx', 'perceraian2019.xlsx', 'perceraian2020.xlsx', 'perceraian2021.xlsx', 'perceraian2022.xlsx']

# Membaca setiap file Excel dan menyimpannya dalam sebuah DataFrame
dfs = [pd.read_excel(file) for file in file_names]
```

```python
# Menggabungkan DataFrame berdasarkan kolom 'Provinsi'
df_combined = pd.concat(dfs, ignore_index=True)

# Menampilkan beberapa baris pertama DataFrame sebelum penjumlahan
print("DataFrame Sebelum Penjumlahan:")
print(df_combined.head())

# Mengelompokkan dan menjumlahkan nilai berdasarkan kolom 'Provinsi'
df_total = df_combined.groupby('Provinsi').sum().reset_index()

# Menampilkan beberapa baris pertama DataFrame setelah penjumlahan
print("\nDataFrame Setelah Penjumlahan:")
print(df_total.head())

# Simpan DataFrame hasil penjumlahan ke file Excel
df_total.to_excel('hasil_gabungan.xlsx', index=False)
```
