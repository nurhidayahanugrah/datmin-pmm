```python
import numpy as np
import pandas as pd
```

```python
# membaca file dataset
dt18 = pd.read_excel("perceraian2018.xlsx")
dt19 = pd.read_excel("perceraian2019.xlsx")
dt20 = pd.read_excel("perceraian2020.xlsx")
dt21 = pd.read_excel("perceraian2021.xlsx")
dt22 = pd.read_excel("perceraian2022.xlsx")
```

```python

# Menggabungkan dataset tahun 2018-2022 berdasarkan kolom 'Provinsi'
df_combined = pd.merge(dt18, dt19, on='Provinsi', how='outer')
df_combined = pd.merge(df_combined, dt20, on='Provinsi', how='outer')
df_combined = pd.merge(df_combined, dt21, on='Provinsi', how='outer')
df_combined = pd.merge(df_combined, dt22, on='Provinsi', how='outer')

# Mengisi nilai NaN dengan 0 (jika ada provinsi yang tidak ada data untuk tahun tertentu)
df_combined = df_combined.fillna(0)

# Mengelompokkan data berdasarkan 'Provinsi' dan menjumlahkan berdasarkan faktor perceraian
df_total = df_combined.groupby('Provinsi').agg({
    'Fakor Perceraian - Zina': 'sum',
    'Fakor Perceraian - Mabuk': 'sum',
    'Fakor Perceraian - Madat': 'sum',
    'Fakor Perceraian - Judi': 'sum',
    'Fakor Perceraian - Meninggalkan Salah satu Pihak': 'sum',
    'Fakor Perceraian - Dihukum Penjara': 'sum',
    'Fakor Perceraian - Poligami': 'sum',
    'Fakor Perceraian - Kekerasan Dalam Rumah Tangga': 'sum',
    'Fakor Perceraian - Cacat Badan': 'sum',
    'Fakor Perceraian - Perselisihan dan Pertengkaran Terus Menerus': 'sum',
    'Fakor Perceraian - Kawin Paksa': 'sum',
    'Fakor Perceraian - Murtad': 'sum',
    'Fakor Perceraian - Ekonomi': 'sum',
    'Fakor Perceraian - Lain-lain': 'sum',
    'Fakor Perceraian - Jumlah': 'sum'
}).reset_index()

# Simpan dataset hasil gabungan
df_total.to_csv('total_perceraian_per_provinsi_by_factor.csv', index=False)
```
